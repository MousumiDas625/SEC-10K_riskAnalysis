```python
#plot a word count of each risk factor
# Count words in each paragraph
def word_count(risk):
    return len(risk.split(" "))

#count words for a list of paragraphs
def word_count_list(lis):
    counts=[]
    for item in lis:
        c=len(item.split(" "))
        counts.append(c)
    return counts    
```
