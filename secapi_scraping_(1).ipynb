```python
#get the ticker list
import numpy as np
import pandas as pd
import os
import json
import requests

headers = {'User-Agent': "mosh25mi@gmail.com"}
# get all companies data
companyTickers = requests.get(
    "https://www.sec.gov/files/company_tickers.json",
    headers=headers
    )
firstEntry = companyTickers.json()['0']

companyData = pd.DataFrame.from_dict(companyTickers.json(),orient='index')

print(companyData.shape)
```

```python
from sec_api import QueryApi
queryApi = QueryApi(api_key="b2612513368fb380c7eb8976321e47dda0a1ae61c4d43d9cf1a7956c37ff5ec9")
```

```python
ticker_list = allticks[:]   #["MSFT", "AAPL", "GOOG"]
query_template = {
    "query": {
        "query_string": {
            "time_zone": "America/New_York"
        }
    },
    "from": "0",
    "size": "9999",
    "sort": [{"filedAt": {"order": "desc"}}]
}

responses1000 = []
all_filings1000 = []

i=0
for ticker in ticker_list:
    query = "ticker:" + ticker + " AND filedAt:[2004-01-01 TO 2023-12-31] AND formType:\"10-K\""
    query_template["query"]["query_string"]["query"] = query
    response = queryApi.get_filings(query_template)
    responses1000.append(response)
    filings = response["filings"]
    all_filings1000.extend(filings)
    print(ticker+" "+str(i))
    i=i+1
```

```python
from sec_api import ExtractorApi
extractorApi = ExtractorApi(api_key="b2612513368fb380c7eb8976321e47dda0a1ae61c4d43d9cf1a7956c37ff5ec9")
```

```python
t=0
all_10k_risks1A=[]
all_10k_risks7=[]
all_10k_risks7A=[]
for i in df_filings1000['linkToFilingDetails'][:5000]:
    item_1_a_text  = extractorApi.get_section(i, '1A', 'text')
    item_7_text = extractorApi.get_section(i, '7', 'text')
    item_7A_text = extractorApi.get_section(i, '7A', 'text')
    all_10k_risks1A.append(item_1_a_text)
    all_10k_risks7.append(item_7_text)
    all_10k_risks7A.append(item_7A_text)
    print(t)
    t=t+1
```

```python
print(df_filings1000.shape)
print(len((all_10k_risks1A)))
print(len((all_10k_risks7)))
print(len((all_10k_risks7A)))
```
